### Step 1: Import Libraries and load to the enviroment variables

```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown,display
```

```python
load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY") 
if OPENAI_API_KEY is None:
    raise Exception("API key is missing.")
else:
   print(OPENAI_API_KEY[:8])
client = OpenAI()
```

Step 2: Call the API

```python
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role":"system", "content":"You are a helpful assitant."},
        {"role":"user", "content":"Next week I'm travelling to Spain"}
    ]
)

reply = response.choices[0].message.content

display(Markdown(reply))
```

```python
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role":"system", "content":"You are a helpful assitant."},
        {"role":"user", "content":"Where am I travelling to next week?"}
    ]
)

reply = response.choices[0].message.content

display(Markdown(reply))
```

```python
messages = [
        {"role":"system", "content":"You are a helpful assitant."},
        {"role":"user", "content":"Next week I'm travelling to Spain"},
        {"role":"assistant", "content":"That sounds exciting! Do you need any tips or information about traveling to Spain? For example, recommendations on places to visit, local customs, language tips, or packing advice?"},
        {"role":"user", "content":"Where am I travelling to next week?"}
    ]

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

reply = response.choices[0].message.content
display(Markdown(reply))

```
