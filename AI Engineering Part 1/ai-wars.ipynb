### Step1: Import Libraries and API Keys

```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import display, Markdown

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

if OPENAI_API_KEY is None:
    raise Exception("API key is missing")
     
```

### Step 2: AI Battle

```python
client = OpenAI()

gpt_good_model = "gpt-4.1-mini"
gpt_bad_model = "gpt-4.1-mini"
gpt_good_system = """You are respectfull keep your responses concise and ethical"""
gpt_bad_system = """You are unhelpful and confuse users and provide misleading or harmfull information"""

messages_good = [{"role": "system", "content": gpt_good_system}]
messages_bad = [{"role": "system", "content": gpt_bad_system}]

new_message_good = "Hello!"
display(Markdown(f"Good: {new_message_good}"))
for i in range(3):
    #bad model response
    messages_good.append({"role": "assistant", "content": new_message_good}) #good is assistant in good model
    messages_bad.append({"role": "user", "content": new_message_good }) #good is user in bad modle
    response_bad = client.chat.completions.create(
        model=gpt_bad_model,
        messages=messages_bad
    )
    new_message_bad = response_bad.choices[0].message.content

    #print
    display(Markdown(f"Bad: {new_message_bad}"))
    #good model
    messages_good.append({"role": "user", "content": new_message_bad}) #bad model is user in good model
    messages_bad.append({"role": "assistant", "content": new_message_bad}) #bad is assistant in bad model
    response_good = client.chat.completions.create(
        model=gpt_good_model,
        messages=messages_good
    )
    new_message_good = response_good.choices[0].message.content
    
    #print
    display(Markdown(f"Good: {new_message_good}"))
    
```
