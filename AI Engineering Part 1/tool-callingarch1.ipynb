```python
pushover_user = os.getenv("PUSHOVER_USER")
pushover_token = os.getenv("PUSHOVER_TOKEN")
pushover_url = "https://api.pushover.net/1/messages.json"
```

```python
#Test Pushover
import requests

def send_notification(message: str):
    payload = {"user": pushover_user, "token": pushover_token, "message": message}
    requests.post(pushover_url, data=payload)
```

```python
send_notification("Hello to myselft, from this AI engineering training")
```

###Step 3: Describe Pushover as an LLM tool

```python
send_notification_function = {
    "name": "send_notification",
    "description": "sends a push notification",
    "parameters":{
        "type": "object",
        "properties":{
            "message":{
                "type": "string",
                "description": "The notification message to send to the user's device"
            }
        },
        "required": ["message"]
}
}
```

###STep 4: Add Pushover to the list of tool for the LLM

```python
tools = [{"type":"function", "function":send_notification_function}]
```

```python

client = OpenAI()
```

###Step 5: Calling the tool from an LLM

```python

client = OpenAI()
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role":"user", "content":"Please send me notificiaton telling me what amazing progress\
         I'm making on the AI Engineering training by superdatascience."}

    ],
    tools=tools
)

#check if model wants to call a tool
message = response.choices[0].message
```

```python
from openai import OpenAI

client = OpenAI()

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role": "user", "content": "Please send me notification telling me what amazing progress\
         I'm making on the AI Engineering training by superdatascience."
        }
    ],
    tools=tools
)

#check if model wants to call a tool
message = response.choices[0].message
```

```python
print(message)
```

```python

if message.tool_calls:
    tool_call = message.tool_calls[0]
    import json
    args = json.loads(tool_call.function.arguments)

    #Actually send the notification
    send_notification(args["message"])
    print(f"Sent notification: {args['message']}")
else:
    print(message.content)
```
