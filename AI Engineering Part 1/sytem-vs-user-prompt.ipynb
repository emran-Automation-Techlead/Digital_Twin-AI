### Step 1: Import Libraries and load to the enviroment variables

```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown,display

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY") 
if OPENAI_API_KEY is None:
    raise Exception("API key is missing.")
else:
   print(OPENAI_API_KEY[:8])

client = OpenAI()

```

###Step 2: Call the API

```python

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[{"role":"user", "content":"HI My name is Imran"}]
)

reply = response.choices[0].message.content
print(reply)
```

###Step 3a : Calling the API with a System Prompt

```python
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role":"system", "content":"you are a helpful assistant."}, ###system prompt
        {"role":"user", "content":"How do I cook pasta?"}
    ]
)
reply = response.choices[0].message.content
#print(reply)
display(Markdown(reply))
```

###Step 3b : Changing the System Prompt Changes the LLM's

```python
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role":"system", "content":"you are a mischievous assistant who always gives incorrect and misleading advice."}, ###system prompt
        {"role":"user", "content":"How do I cook pasta?"}
    ]
)
reply = response.choices[0].message.content
#print(reply)
display(Markdown(reply))
```
