### Step1: Import Libraries and API Keys

```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import display, Markdown
from openai import OpenAI
import gradio as gr

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

if OPENAI_API_KEY is None:
    raise Exception("API key is missing")

     
```

###Step 2: Simple UI with AI

```python


def respond_ai(message, history):
    messages = [{"role": "system", "content": "You are helpful assistant  replies"}] + history + [{"role": "user", "content": message}]
    client = OpenAI(api_key=OPENAI_API_KEY)
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages
    )
    reply = response.choices[0].message.content
    return  reply


```

```python
gr.ChatInterface(fn=respond_ai).launch(inbrowser=True)
```

### Step 5: Dynamic Context Injection

```python
Topic_Context = {
    "September 2018 and January 2019": "***Imran Joined ITILv3 certification training.***",
    "Guava": "Imran enjoys juice.",
    "snack": "Imran enjoys snack with sover sweet added guava",
    "Brinjal": "Imran least liked food vegitable is brinjal.But he likes it air fried."

}
```

```python


def respond_ai(message, history):
      #Inject Dynamic Context based on keywords in the message
    system_message_enhanced = system_message
    for keyword, context in Topic_Context.items():
        if keyword in message.lower():
            system_message_enhanced += "\n\n" + context
          
    #As usual

    messages = [{"role": "system", "content": system_message_enhanced}] + history + [{"role": "user", "content": message}]
    client = OpenAI(api_key=OPENAI_API_KEY)
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages
    )
    reply = response.choices[0].message.content
    return  reply
gr.ChatInterface(fn=respond_ai).launch(inbrowser=True)

```
